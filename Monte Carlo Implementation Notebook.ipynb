```python
from monte_carlo import monte_carlo_sim
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Loading the weather data

data = pd.read_csv('Mean_Temp_IMD_2017.csv')
```

```python
data
```

```python
sim_jan = data.iloc[96:106,1]
test_sequence = data.iloc[106:116,1].reset_index().iloc[:,1]
labels = list(data.iloc[106:116,0])
x = np.arange(0,10)
```

```python
mu_t = np.sum(sim_jan)/len(sim_jan)
tau_t = 1 / np.var(sim_jan)
sim_vals = monte_carlo_sim(sim_jan, mu_t,tau_t, 1 , 25, 40.5 , 10)
```

```python
plt.figure(figsize = (8,5))
plt.plot(sim_vals , color = 'red' , label = 'Simulated' , lw = '3')
plt.plot(test_sequence , color = 'blue' , label = 'observed' , lw = '3')
plt.xticks(x,labels)
plt.ylabel('Temperature - degree celsius')
plt.xlabel('Years')
plt.legend()
plt.show()
```
